## Climate Adaptation and Security in the Sahel and Horn of Africa

### Set up the required libraries and load the dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA

file_path = "data/Adenium_Dataset.xlsx"
df = pd.read_excel(file_path)
dependent_variables = 'conflict_count'
```

### Correlation Matrix

```python
check = ['oda_total_net',
       'odf_total_net', 'female_pop', 'male_pop', 'people_in_need',
       'INFORM_risk', 'military_spending',
       'ipc_food_insecurity', 'rule_of_law', 'gender_devel', 'mpi',
       'nd_readiness', 'nd_vulnerability', 'foreign_invest',
       'rural_population', 'fpi', 'lpi', 'rfh_avg', 'c_drought', 'c_flood',
       'c_storm']
corr_matrix = df[check].corr()

print("Correlation Matrix:")
display(corr_matrix)

plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Correlation Matrix - Independent Variables of Climate-Conflict-Adaptation", fontsize = 12)
plt.savefig('visual/corr_matrix.png', dpi=300, bbox_inches='tight', format='png')
plt.show()
```

### Variables Setting

```python
independent_variables = ['odf_total_net', 'female_pop', 'people_in_need',
       'INFORM_risk', 'military_spending',
       'ipc_food_insecurity', 'gender_devel', 'mpi',
       'nd_readiness', 'foreign_invest',
       'rural_population', 'lpi', 'rfh_avg', 'c_drought', 'c_flood',
       'c_storm']
```

### Principal Components Analysis

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), independent_variables)
    ]
)

X = df[independent_variables]
X_preprocessed = preprocessor.fit_transform(X)
y = df[dependent_variables]

X_preprocessed_df = pd.DataFrame(X_preprocessed, columns=independent_variables)

n_components = 9
pca = PCA(n_components=n_components)

X_pca = pca.fit_transform(X_preprocessed_df)

pca_columns = [f'PC{i+1}' for i in range(X_pca.shape[1])]
X_pca_df = pd.DataFrame(X_pca, columns=pca_columns)

explained_variance = pca.explained_variance_ratio_
print("Explained variance by each component:", explained_variance)

total_explained_variance = explained_variance.sum()
print("Total explained variance:", total_explained_variance)
```

```python
loadings = pd.DataFrame(
    pca.components_.T, 
    columns=[f'PC{i+1}' for i in range(n_components)],
    index=X_preprocessed_df.columns
)

threshold = 0.3

important_vars = {}

for pc in loadings.columns:
    vars_above_threshold = loadings[pc][loadings[pc].abs() >= threshold].sort_values(ascending=False)
    important_vars[pc] = vars_above_threshold

important_vars_df = pd.concat(important_vars, axis=1)
important_vars_df.fillna('', inplace=True)

print("Variables for each PC (|loading| >= 0.3):")
display(important_vars_df)
```

```python
bodhi_blue = (0.0745, 0.220, 0.396)
bodhi_grey = (0.247, 0.29, 0.322)
bodhi_primary_1 = (0.239, 0.38, 0.553)
bodhi_secondary = (0.133, 0.098, 0.42)
bodhi_tertiary = (0.047, 0.396, 0.298)
bodhi_complement = (0.604, 0.396, 0.071)

bodhi_palette = [bodhi_complement, bodhi_blue, bodhi_tertiary,
                 bodhi_primary_1, bodhi_grey, bodhi_secondary]

def plot_horizontal_bar(data_dict, PC, filename="pc1_plot.png", folder="visual", title="PC1 Loadings"):

    os.makedirs(folder, exist_ok=True)
    
    labels = list(data_dict.keys())
    values = list(data_dict.values())
    
    colors = bodhi_palette[:len(labels)]
    
    fig, ax = plt.subplots(figsize=(8, 5), dpi=300)
    
    bars = ax.barh(labels, values, color=colors)
    
    ax.axvline(0, color='black', linewidth=1)
    
    for bar, value in zip(bars, values):
        ax.text(
            bar.get_width()/2,  
            bar.get_y() + bar.get_height()/2,
            f"{value:.4f}",  
            ha='center', va='center', color='white', fontsize=13, fontweight='bold'
        )
    
    ax.set_xlabel("")
    ax.tick_params(axis='y', labelsize=15)
    ax.set_title(title, fontsize = 16)
    
    plt.tight_layout()
    
    save_path = os.path.join(folder, filename)
    plt.savefig(save_path, dpi=300, bbox_inches='tight', format='png')
    plt.close()

data1 = {
    "INFORM_risk": 0.4520,
    "people_in_need": 0.3784,
    "rural_population": 0.3705,
    "foreign_invest": -0.3614
}

plot_horizontal_bar(data1, PC ='PC1', filename="pc1_loadings.png")

data2 = {
    "gender_devel": 0.4328,
    "odf_total_net": 0.3437,
    "female_pop": 0.3073,
    "military_spending": -0.4018,
    "mpi": -0.4227
}

data3 = {
    "foreign_invest": 0.4614,
    "nd_readiness": 0.3029,
    "rfh_avg": 0.3750,
    "c_flood": 0.4741
}

data4 = {
    "c_drought": 0.6323,
    "c_storm": -0.3044
}

plot_horizontal_bar(data1, PC ='PC1', filename="pc1_loadings.png")
plot_horizontal_bar(data2, PC ='PC2', filename="pc2_loadings.png", title="PC2 Loadings")
plot_horizontal_bar(data3, PC ='PC4', filename="pc4_loadings.png", title="PC4 Loadings")
plot_horizontal_bar(data4, PC ='PC5', filename="pc5_loadings.png", title="PC5 Loadings")
```

```python
eigenvalues = pca.explained_variance_
print("eigenvalues of Each PCs:")
for i, ev in enumerate(eigenvalues, 1):
    print(f"PC{i}: {ev:.4f}")
```

### Principal Components Regression - UCDP

```python
import statsmodels.api as sm

X_sm = sm.add_constant(X_pca_df.reset_index(drop=True))
y = df['conflict_count'].reset_index(drop=True)

model = sm.OLS(y, X_sm).fit()

print(model.summary())
```
